```python
air1=(air.iloc[:,1:])
air1
```

```python
plt.figure(figsize=(15,12));plt.title('HIERARCHY DENDROGRAM');plt.xlabel('Index');plt.ylabel('Distance')
dendograms=sch.dendrogram(sch.linkage(air1_norm,"complete"))
plt.show()
```

```python
air=pd.read_excel('/content/drive/MyDrive/Assignment-Data Science/7.Clustering /data/EastWestAirlines.xlsx',sheet_name='data')
```

```python
# Normalization function
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
df_norm = norm_func(air.iloc[:,1:])


df_norm.head(10)
```

```python
# Selecting 5 clusters from the above
model=KMeans(n_clusters=5)
model.fit(df_norm)

model.labels_
```

```python
md=pd.Series(model.labels_)  # converting numpy array into pandas series object
air['clust']=md
df_norm.head()
```

```python
air.iloc[:,1:7].groupby(air.clust).mean()
```

```python
air.head()
```

```python
Wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i,random_state=2)
    kmeans.fit(df_norm)
    Wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11),Wcss)
plt.title('Elbow Graph')
plt.xlabel('Number of clusters')
plt.ylabel('Wcss')
plt.show()
```

# DBSCAN

```python
# Import .csv file and convert it to a DataFrame object
air=pd.read_excel('/content/drive/MyDrive/Assignment-Data Science/7.Clustering /data/EastWestAirlines.xlsx',sheet_name='data')

print(air.head())
```

```python
print(air.info())
```

```python
air.drop(['ID#','Award?'],axis=1,inplace=True)
```

```python
air
```

```python
array=air.values
```

```python
array
```

```python
stscaler = StandardScaler().fit(array)
X = stscaler.transform(array)
```

```python
X
```

```python
dbscan = DBSCAN(eps=1, min_samples=12)
dbscan.fit(X)
```

```python
#Noisy samples are given the label -1.
dbscan.labels_
```

```python
cl=pd.DataFrame(dbscan.labels_,columns=['cluster'])
```

```python
cl
```

```python
pd.concat([air,cl],axis=1)
```
